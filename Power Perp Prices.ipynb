```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sbn
import pandas as pd
from itertools import cycle, islice
```

```python
def power_perp_price(spot, time, vol, drift, power):
    price = spot ** power * np.exp((power - 1) * (drift + power / 2 * vol ** 2) * time)
    return price


def everlasting_power_perp_price(spot, funding_period, vol, drift, power):
    price = spot ** power * (1 / (2* np.exp(-funding_period/2 * (power - 1) * (2 * drift + power * vol **2)) - 1))
    return price
```

```python
vol = 1.5
drift = 0
funding_period = 1/365
spot = np.arange(1000., 4000., 50.)
```

```python
power = 2
payoff = spot ** power / spot[0]**(power-1)
price = everlasting_power_perp_price(spot, funding_period, vol, drift, power) / spot[0]**(power-1)

data = pd.DataFrame({'spot ETH': spot, f'ETH^{power}': payoff, f'ETH^{power} perp price': price}).set_index('spot ETH')
figure(figsize=(12, 9), dpi=80)
ax = sbn.lineplot(data=data)
ax.set_ylabel("value")
ax.set_title(f"Normalized ETH^{power} Perpetual Price vs. Spot Price ({int(vol*100)} vol)")
```
